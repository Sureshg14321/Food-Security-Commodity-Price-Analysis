```python
num_cols = commodity.select_dtypes(include=np.number).columns

commodity[num_cols] = commodity[num_cols].fillna(
    commodity[num_cols].median()
)
```

```python
num_cols = crop.select_dtypes(include=np.number).columns

crop[num_cols] = crop[num_cols].fillna(
    crop[num_cols].median()
)
```

```python
num_cols = fsi.select_dtypes(include=np.number).columns

fsi[num_cols] = fsi[num_cols].fillna(
    fsi[num_cols].median()
)
```

```python
num_cols = panel.select_dtypes(include=np.number).columns

panel[num_cols] = panel[num_cols].fillna(
    panel[num_cols].median()
)
```

```python
avg_food_price_index = panel['ffpi_overall'].mean()

print("Average Food Price Index:", avg_food_price_index)
```

```python
avg_wheat_price = panel['wheat_usd_mt'].mean()

print("Average Wheat Price:", avg_wheat_price)
```

```python
avg_maize_price = panel['maize_usd_mt'].mean()

print("Average Maize Price:", avg_maize_price)
```

```python
avg_fertilizer_index = panel['fertiliser_composite_idx'].mean()

print("Average Fertilizer Index:", avg_fertilizer_index)
```

```python
avg_crop_yield = crop['total_staple_production_mt'].mean()

print("Average Crop Yield:", avg_crop_yield)
```

```python
avg_mom_growth = panel['ffpi_overall_mom_pct'].mean()

print("Average MOM Growth %:", avg_mom_growth)
```

```python
panel = panel.sort_values('date')

panel['ffpi_overall_yoy_pct'] = (
    panel['ffpi_overall']
    .pct_change(periods=12)
) * 100
```

```python
panel[['date','ffpi_overall','ffpi_overall_yoy_pct']].head(20)
```

```python
avg_yoy_growth = panel['ffpi_overall_yoy_pct'].mean()

print(avg_yoy_growth)
```

```python
crisis_count = panel['ffpi_crisis_flag'].sum()

print("Crisis Count:", crisis_count)
```

```python
supply_stress_count = panel['supply_stress_flag'].sum()

print("Supply Stress Count:", supply_stress_count)
```

```python
kpi_summary = pd.DataFrame({
    'Metric':[
        'Average Food Price Index',
        'Average Wheat Price',
        'Average Maize Price',
        'Average Fertilizer Index',
        'Average Crop Yield',
        'Average MOM Growth %',
        'Average YOY Growth %',
        'Crisis Count',
        'Supply Stress Count'
    ],
    'Value':[
        avg_food_price_index,
        avg_wheat_price,
        avg_maize_price,
        avg_fertilizer_index,
        avg_crop_yield,
        avg_mom_growth,
        avg_yoy_growth,
        crisis_count,
        supply_stress_count
    ]
})

kpi_summary
```
